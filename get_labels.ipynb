```python
import os
import json
import pandas as pd
file_names = []
for file_name in os.listdir('./数据集id'):
    if('v4' in file_name and 'heal' in file_name):
        file_names.append('./数据集id/'+file_name)
```

```python
file_names
```

```python

for file_name in file_names:
    dfs = []
    with open(file = file_name,encoding='utf-8') as f:
        label_name = './标签/label_'+file_name.split('/')[-1].split('.')[0]+'.csv'
        index = 0
        print(file_name)
        with open(file = label_name,encoding='utf-8',mode='a') as s:
            for line in f:
                obj_new = {}
                obj = (json.loads(line))
                obj_new['uuid'] = obj['uuid']
                obj_new['post_url'] = obj['post_url']
                obj_new['post_time'] = obj['post_time']
                obj_new['post_user_name'] = obj['post_user_name']
                obj_new['post_text'] = obj['post_text']
                obj_new['post_reaction_num'] = obj['post_reaction_num']
                obj_new['post_share_comment'] = obj['post_share_comment']
                obj_new['label'] = ''
                dfs.append(pd.DataFrame(obj_new,index=[index]))
                index += 1
        pd.concat(dfs).to_csv(label_name)
                
```
